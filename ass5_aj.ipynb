```python
import numpy as np
import pandas as pd
```

```python
#to read data
df=pd.read_csv("Social_Network_Ads.csv")
df
```

```python
#drop the userid and gender because its non numeric
x=df.drop(['User ID','Gender'],axis=1)
x
```

```python
x.dtypes
```

```python

y=df['Purchased']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=9) 
#in random state we can take any value, when random state is changed the accuracy will also change
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
acc=accuracy_score(y_test,y_predict)
print(acc)
```

```python
cm=confusion_matrix(y_predict,y_test)
cm
```

```python
prec_score=precision_score(y_test,y_predict)
print(prec_score)
```

```python
rec_score=recall_score(y_test,y_predict)
rec_score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
