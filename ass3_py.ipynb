```python
def display_statistics(species_data,species_name):
    nc=["sepal_length",  "sepal_width",  "petal_length",  "petal_width"]
    print(f"\n\nstatistics for: {species_name}")
    print("mean: ")
    print(species_data[nc].mean())
    print("median: ")
    print(species_data[nc].median())
    print("max: ")
    print(species_data[nc].max())
    print("standard deviation: ")
    print(species_data[nc].std())
    print("min: ")
    print(species_data[nc].min())
    print("25%: ")
    print(species_data[nc].quantile(0.25))
    print("50%: ")
    print( species_data[nc].quantile(0.50))
    print("75%: ")
    print(species_data[nc].quantile(0.75) )
    
setosa_data=iris_data[iris_data['species']=='setosa']
display_statistics(setosa_data,'Iris-setosa')

versicolor_data=iris_data[iris_data['species']=='versicolor']
display_statistics(versicolor_data,'Iris-versicolor')

virginica_data=iris_data[iris_data['species']=='virginica']
display_statistics(virginica_data,'Iris-virginica')   
```

```python
import pandas as pd

def calc_mean(data):
    if len(data) == 0:
        return None
    else:
        return sum(data) / len(data)

def calc_std_dev(data, mean):
    if len(data) == 0:
        return None
    else:
        squarediff = sum((x - mean)**2 for x in data)
        return (squarediff / (len(data) - 1))**0.5

def per_calc(data, percentile):
    sorted_data = sorted(data)
    index = int(percentile * len(data))
    return sorted_data[index]

def display_statistics(species_data, species_name):
    nc = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
    
    print(f"\n\nstatistics for: {species_name}")
    
    print("\nmean: ")
    mean_val = [calc_mean(species_data[col]) for col in nc]
    print(pd.Series(mean_val, index=nc))
    
    print("\nstandard deviation: ")
    std_val = [calc_std_dev(species_data[col], mean_val[i]) for i, col in enumerate(nc)]
    print(pd.Series(std_val, index=nc))
    
    percent = [0.25, 0.50, 0.75]
    for col in nc:
        print("\n",col)
        for percentile in percent:
            print(f"{percentile} percentile: {per_calc(species_data[col], percentile)}")
# Assuming you have 'iris_data' DataFrame available
iris_data=sns.load_dataset('iris')  # Replace 'your_iris_data.csv' with the actual file name

setosa_data = iris_data[iris_data['species'] == 'setosa']
display_statistics(setosa_data, 'Iris-setosa')

versicolor_data = iris_data[iris_data['species'] == 'versicolor']
display_statistics(versicolor_data, 'Iris-versicolor')

virginica_data = iris_data[iris_data['species'] == 'virginica']
display_statistics(virginica_data, 'Iris-virginica')
```
