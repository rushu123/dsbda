```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',data=df)
```

```python
sns.displot(x='age',data=df,col='embark_town',bins=10,hue='sex',multiple='dodge')
```

```python
sns.barplot(x='sex',y='age',data=df,hue='sex',errorbar=None) # instead of errorbar=None, ci=none also works
```

```python
sns.barplot(x='pclass',y='age',hue='embark_town',data=df,errorbar=None)
```

```python
sns.barplot(x='pclass',y='fare',hue='embark_town',data=df,errorbar=None)
```

```python
sns.countplot(x='pclass',data=df,hue='sex')
```

```python
sns.countplot(x='pclass',data=df,hue='embark_town')
```

```python
sns.countplot(x='pclass',data=df,hue='survived')
```

```python
sns.countplot(x='survived',data=df,hue='embark_town')
```

```python
sns.catplot(x='embark_town',kind='count',data=df,hue='survived',col='pclass')
```
