```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('A1.csv')

# Extract relevant columns
grades = df[['Roll No', 'Subject 1', 'Subject 2', 'Subject 3', 'Subject 4', 'Attendance']]

# Plot Histograms
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Histograms for Subject Grades and Attendance', y=0.95)
for i, col in enumerate(grades.columns[1:]):  # Exclude 'Roll No'
    ax = axes.flatten()[i]
    ax.hist(grades[col], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f'{col} Histogram')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Q-Q plot
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Q-Q Plots for Subject Grades and Attendance', y=0.98)
for i, col in enumerate(grades.columns[1:]):  # Exclude 'Roll No'
    ax = axes.flatten()[i]
    sorted_data = np.sort(grades[col])
    theoretical_quantiles = np.percentile(sorted_data, np.linspace(0, 100, len(sorted_data)))
    ax.plot(theoretical_quantiles, np.sort(grades[col]), marker='o', linestyle='None', color='blue')
    ax.plot([np.min(sorted_data), np.max(sorted_data)], [np.min(sorted_data), np.max(sorted_data)], linestyle='--', color='red')
    ax.set_title(f'{col} Q-Q Plot')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# KDE plot
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Kernel Density Estimation Plots for Subject Grades and Attendance', y=0.98)
for i, col in enumerate(grades.columns[1:]):  # Exclude 'Roll No'
    ax = axes.flatten()[i]
    sns.kdeplot(grades[col], ax=ax, fill=True, color='skyblue', linewidth=2)
    ax.set_title(f'{col} KDE Plot')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Skewness and z-score
skewness = grades.skew()
z_scores = (grades - grades.mean()) / grades.std()

# Display results
print("Skewness:")
print(skewness)
print("\nZ-Scores:")
print(z_scores)
```
